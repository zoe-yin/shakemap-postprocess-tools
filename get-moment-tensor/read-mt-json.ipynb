```python
import json
from obspy.imaging.beachball import MomentTensor, mt2plane, aux_plane


file = '/Users/hyin/shakemap_profiles/default/data/us6000ah9t/us6000ah9t_tensor.json'
# file = '/Users/hyin/shakemap_profiles/default/data/ci38457511/ci38457511_tensor.json'
```

```python
with open(file) as f:
    d = json.load(f)

props = d["properties"]

# Extract tensor components (convert strings to float)
mrr = float(props["tensor-mrr"])
mtt = float(props["tensor-mtt"])
mpp = float(props["tensor-mpp"])
mrt = float(props["tensor-mrt"])
mrp = float(props["tensor-mrp"])
mtp = float(props["tensor-mtp"])

# Build ObsPy MomentTensor
# Order: Mrr, Mtt, Mpp, Mrt, Mrp, Mtp, random exponent to make the class happy
mt = MomentTensor(mrr, mtt, mpp, mrt, mrp, mtp, 10)
# mt = MomentTensor(1, 1, 0, 0, 0, -1, 12)

# # Compute nodal planes
np1 = mt2plane(mt)
np2 = aux_plane(np1.strike,np1.dip, np1.rake)
print(np2)


print("Nodal Plane 1:")
print(f"  Strike: {np1.strike:.2f}")
print(f"  Dip:    {np1.dip:.2f}")
print(f"  Rake:   {np1.rake:.2f}")

print("Nodal Plane 2:")
print(f"  Strike: {np2[0]}")
print(f"  Dip:    {np2[1]}")
print(f"  Rake:   {np2[2]}")


if d['source'] == 'ci':
    print("Reading California MT solution")
if d['source'] == 'us':
    print("Reading US MT solution")
```
