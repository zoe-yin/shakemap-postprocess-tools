```python
import os
import json
import pandas as pd
import numpy as np

input_excel = "/Users/hyin/usgs_mendenhall/ffsimmer/catalog/catalog-development/7.0events_since_2018_PAGER-filt.xlsx"
output_excel = "/Users/hyin/usgs_mendenhall/ffsimmer/catalog/catalog-development/7.0events_since_2018_PAGER-filt_strec.xlsx"

# Read spreadsheet
df = pd.read_excel(input_excel)

def extract_values_from_json(eventid):
    """Open JSON file and extract desired fields."""
    json_directory=eventpath=f"/Users/hyin/shakemap_profiles/default/data/{eventid}/"
    json_path = os.path.join(json_directory, f"shakemap_reproduction/strec_results.json")
    print(json_path)
    # Keys to extract
    json_keys = [
        "TectonicRegion",
        "FocalMechanism",
        "ProbabilityActive",
        "ProbabilityStable",
        "ProbabilitySubduction",
        "ProbabilityVolcanic",
        "ProbabilitySubductionCrustal",
        "ProbabilitySubductionInterface",
        "ProbabilitySubductionIntraslab",
        "ProbabilityStableShallow",
        "ProbabilityVolcanicShallow",
        "ProbabilityActiveShallow",
        "ProbabilityActiveDeep"
    ]

    if not os.path.exists(json_path):
        print("No JSON file found")
        return {key: np.nan for key in json_keys}

    with open(json_path, "r") as f:
        data = json.load(f)

    # Direct key access (flat JSON)
    extracted = {}
    for key in json_keys:
        value = data.get(key, np.nan)

        # Handle NaN values coming from JSON (if stored as string or null)
        if value is None:
            value = np.nan

        extracted[key] = value

    return extracted


# Extract JSON data for each row
new_columns = df["eventid"].apply(extract_values_from_json)

# Convert list of dicts into dataframe
new_df = pd.DataFrame(list(new_columns))

# Combine with original
df_out = pd.concat([df, new_df], axis=1)

# Write new Excel file
df_out.to_excel(output_excel, index=False)
print(f"New file written to: {output_excel}")
```
